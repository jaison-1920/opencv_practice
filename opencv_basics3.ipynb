```python
import os
import cv2
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from zipfile import ZipFile
from urllib.request import urlretrieve

matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)

%matplotlib inline
```

```python
def download_and_unzip(url, save_path):
    print(f"Downloading and extracting assests....", end="")

    # Downloading zip file using urllib package.
    urlretrieve(url, save_path)

    try:
        # Extracting zip file using the zipfile package.
        with ZipFile(save_path) as z:
            # Extract ZIP file contents in the same directory.
            z.extractall(os.path.split(save_path)[0])

        print("Done")

    except Exception as e:
        print("\nInvalid file.", e)
```

```python
URL = r"https://www.dropbox.com/s/48hboi1m4crv1tl/opencv_bootcamp_assets_NB3.zip?dl=1"

asset_zip_path = os.path.join(os.getcwd(), f"opencv_bootcamp_assets_NB3.zip")

# Download if assest ZIP does not exists.
if not os.path.exists(asset_zip_path):
    download_and_unzip(URL, asset_zip_path)
```

# Annotating Images
#### Draw lines
#### Draw circles
#### Draw rectangles
#### Add text

```python
apollo_img = cv2.imread('Apollo_11_Launch.jpg',1)
apollo_img=apollo_img[:,:,::-1]
plt.imshow(apollo_img)
```

# Draw a line

```python
img_line = apollo_img.copy()
# cv2.line(src,point 1,point 2,color,thickness(optional),linetype(optional))
cv2.line(img_line,(100,50),(400,50),(255,255,0))
plt.imshow(img_line)
```

# Draw a circle

```python
# cv2.circle(img, center, radius, color[, thickness[, lineType[, shift]]])
img_circle = apollo_img.copy()
cv2.circle(img_circle, (1000,500),100,(0,255,255), thickness=5,lineType=cv2.LINE_AA)
plt.imshow(img_circle)
```

# Draw a rectangle

```python
# cv2.rectangle(img, top-left vertex, ryt-botto vertex, color, thickness, lineType,)
img_rect = apollo_img.copy()
cv2.rectangle(img_rect,(500,50),(800,650),(255,255,0),thickness=5,lineType=cv2.LINE_AA)
plt.imshow(img_rect)
```

# Add Text

```python
#cv2.putText(img, text, org(Bottom-left corner of the text string in the image), fontFace(font type), fontScale, color[ thickness lineType bottomLeftOrigin)]
img_text = apollo_img.copy()
text = 'Apollo 11'
org = (450,680)
fontFace=cv2.FONT_HERSHEY_COMPLEX
fontScale=2
color = (200,200,0)
cv2.putText(img_text,text,org,fontFace,fontScale,color)
plt.imshow(img_text)
```
